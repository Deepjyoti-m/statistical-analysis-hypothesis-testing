```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\data\raw\ecommerce_data.csv")
print(df.shape)
df.head()
```

```python
print(df.describe())
print("\nnull values:\n", df.isnull().sum())
print("\nvalue counts - Satisfaction:\n", df["Satisfaction Level"].value_counts())
print("\nvalue counts - Membership:\n", df["Membership Type"].value_counts())
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df["Total Spend"], bins=30, color="steelblue", edgecolor="white")
ax.set(title="Distribution of Total Spend", xlabel="Total Spend ($)", ylabel="Count")
plt.tight_layout()
plt.savefig(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\reports\spend_distribution.png")
plt.show()
```

```python
gender_spend = df.groupby("Gender")["Total Spend"].mean().reset_index()

fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(gender_spend["Gender"], gender_spend["Total Spend"], color=["steelblue", "coral"])
ax.set(title="Average Spend by Gender", xlabel="Gender", ylabel="Avg Spend ($)")
plt.tight_layout()
plt.savefig(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\reports\spend_by_gender.png")
plt.show()
```

```python
membership_spend = df.groupby("Membership Type")["Total Spend"].mean().reset_index()
membership_spend = membership_spend.sort_values("Total Spend", ascending=False)

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(membership_spend["Membership Type"], membership_spend["Total Spend"], color=["gold", "silver", "peru"])
ax.set(title="Average Spend by Membership Type", xlabel="Membership Type", ylabel="Avg Spend ($)")
plt.tight_layout()
plt.savefig(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\reports\spend_by_membership.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df["Average Rating"], bins=20, color="mediumpurple", edgecolor="white")
ax.set(title="Distribution of Average Rating", xlabel="Rating", ylabel="Count")
plt.tight_layout()
plt.savefig(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\reports\rating_distribution.png")
plt.show()
```

```python
satisfaction = df["Satisfaction Level"].value_counts().reset_index()
satisfaction.columns = ["Satisfaction", "Count"]

fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(satisfaction["Satisfaction"], satisfaction["Count"], color=["seagreen", "steelblue", "coral"])
ax.set(title="Customer Satisfaction Breakdown", xlabel="Satisfaction Level", ylabel="Count")
plt.tight_layout()
plt.savefig(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\reports\satisfaction_breakdown.png")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(df["Age"], df["Total Spend"], color="steelblue", alpha=0.5)
ax.set(title="Age vs Total Spend", xlabel="Age", ylabel="Total Spend ($)")
plt.tight_layout()
plt.savefig(r"C:\Users\deepj\Documents\statistical-analysis-hypothesis-testing\reports\age_vs_spend.png")
plt.show()
```
